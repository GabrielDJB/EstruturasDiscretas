### Primeira Quest√£o - Obter o coeficiente, baseado na prova do enunciado.

```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
from CPUtimer import *

# Fun√ß√£o de Coeficiente:
#
# Descri√ß√£o: Obt√©m o valor do coeficiente procurado no teorema.
#
# Par√¢metros: x - Valor de x da express√£o
#             y - Valor de y da express√£o
#             n - Valor de n da express√£o
#
# Retorno: q - Coeficiente da divis√£o de x^n - y^n por x - y.

def coeficiente(x, y, n):
    #Checa se n √© positivo e maior que zero
    if(n <= 0):
        print("Parametros invalidos, n menor ou igual a zero!\n")
        return -1
    #Checa se estamos trabalhando com o caso base
    if(n == 1):
        return 1
    #CASO BASE: q1 = 1
    q = 1
    #Ser√° um m√©todo iterativo, processando at√© n
    for i in range(0, n-1):
        q = m.pow(x, (i+1)) + y * q
    return q

# Fun√ß√£o de Teste:
#
# Descri√ß√£o: Exibe cada caso de testes, tendo chamado a fun√ß√£o que obt√©m o coeficiente.
#
# Par√¢metros: x - Valor de x da express√£o
#             y - Valor de y da express√£o
#             n - Valor de n da express√£o
#             i - N√∫mero do caso de teste
#
# Retorno: Imprime no console as informa√ß√µes sobre o caso de teste.

def teste(x,y,n,i,timer):
    print("CASO " + str(i) + "\n")
    
    v1 = m.pow(x,n) - m.pow(y,n)
    v2 = x-y
    timer.start()
    q = coeficiente(x,y,n)
    timer.stop()
    v3 = q * v2
    
    if(q == -1):
        return
    
    print("Valor de X: " + str(x))
    print("Valor de Y: " + str(y))
    print("Valor de N: " + str(n) + "\n")
    print("Valor de x^n - y^n: " + str(v1))
    print("Valor de x - y: " + str(v2))
    print("Valor de qk * (x - y): " + str(v3))
    print("Coeficiente qk: " + str(q) + "\n")
    

timer = CPUTimer(0)
times = np.zeros(10)
n = np.zeros(10)
    
#Primeiro caso
teste(150, 71, 2, 1, timer)
times[0] = timer.get_time(reference = "total", unit = "seconds")
timer.lap(start_stopped = False)
n[0] = 2 

#Segundo caso
teste(5, 3, 1, 2, timer)
times[1] = timer.get_time(reference = "total", unit = "seconds")
timer.lap(start_stopped = False)
n[1] = 1

#Terceiro caso
teste(6, 4, -1, 3, timer)
times[2] = timer.get_time(reference = "total", unit = "seconds")
timer.lap(start_stopped = False)
n[2] = -1

#Quarto caso - A partir daqui varia o n, para ver na fun√ß√£o
teste(145, 5, 3, 4, timer)
times[3] = timer.get_time(reference = "total", unit = "seconds")
timer.lap(start_stopped = False)
n[3] = 3

#Quinto caso
teste(145, 5, 6, 5, timer)
times[4] = timer.get_time(reference = "total", unit = "seconds")
timer.lap(start_stopped = False)
n[4] = 5

#Sexto caso
teste(145, 5, 10, 5, timer)
times[5] = timer.get_time(reference = "total", unit = "seconds")
timer.lap(start_stopped = False)
n[5] = 10

#S√©timo caso
teste(145, 5, 20, 5, timer)
times[6] = timer.get_time(reference = "total", unit = "seconds")
timer.lap(start_stopped = False)
n[6] = 20

#Oitavo caso
teste(145, 5, 30, 5, timer)
times[7] = timer.get_time(reference = "total", unit = "seconds")
timer.lap(start_stopped = False)
n[7] = 30

#Nono caso
teste(145, 5, 40, 5, timer)
times[8] = timer.get_time(reference = "total", unit = "seconds")
timer.lap(start_stopped = False)
n[8] = 40

#Decimo caso
teste(145, 5, 50, 5, timer)
times[9] = timer.get_time(reference = "total", unit = "seconds")
timer.lap(start_stopped = False)
n[9] = 50

#Impress√£o dos Tempos
print(times)
plt.plot(n, times, 'ro')
plt.xlabel("Valor de n")
plt.ylabel("Tempos de execu√ß√£o")
plt.show()

# Conclus√£o:
#
# Com a varia√ß√£o de n podemos ver que h√° uma progress√£o linear temporal.
# Ou seja, a complexidade em fun√ß√£o de n √© linear.
```
